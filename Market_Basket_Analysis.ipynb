```python
#trans_list = []
#for i in range(0, len(data)):
   # trans_list.append([str(data.values[i,j]) for j in range(0, len(data.columns))])
data['transaction'] = data[data.columns[0:]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1
)
```

```python
print(data['transaction'])
```

```python
transactions = data['transaction'].apply(lambda t: t.split(','))# Convert DataFrame into list of strings.transactions = list(transactions)
transactions = list(transactions)
```

```python
transactions
```

```python
from itertools import permutations
# Extract unique items.
flattened = [item for transaction in transactions for item in transaction]
items = list(set(flattened))

# Compute and print rules.
rules = list(permutations(items, 2))
print(rules)
```
